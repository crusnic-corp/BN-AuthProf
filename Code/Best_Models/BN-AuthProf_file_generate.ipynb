```python
data = []
train_data_trial = []
dev_data_trial = []
test_data_trial = []

for i in range(0,300):
  with open('dataset/{0}.txt'.format(i+1),'r', encoding='utf-8') as f:
    for line in f:
      data.append(line)

  length = len(data)
  eighty_percent = math.floor( length * 0.8 )
  twenty_percent = length - eighty_percent
  ten_percent = math.floor(twenty_percent/2)

  train_data_trial.append(data[:eighty_percent])

  dev_data_trial.append(data[eighty_percent:eighty_percent+ten_percent])

  test_data_trial.append(data[eighty_percent+ten_percent:])

  data.clear()
```

```python
for i in range(0,300):
  for line_train in train_data_trial[i]:
    with open('BN-AuthProf_Dataset/Exp_Dataset/Age/Train/{0}.txt'.format(i+1), "w+") as f:
      for line_train in train_data_trial[i]:
        f.write(str(line_train))
    with open('BN-AuthProf_Dataset/Exp_Dataset/Gender/Train/{0}.txt'.format(i+1), "w+") as f:
      for line_train in train_data_trial[i]:
        f.write(str(line_train))
```

```python
for i in range(0,300):
  for line_dev in dev_data_trial[i]:
    with open('BN-AuthProf_Dataset/Exp_Dataset/Age/Dev/{0}.txt'.format(i+1), "w+") as f:
      for line_dev in dev_data_trial[i]:
        f.write(str(line_dev))
    with open('BN-AuthProf_Dataset/Exp_Dataset/Gender/Dev/{0}.txt'.format(i+1), "w+") as f:
      for line_dev in dev_data_trial[i]:
        f.write(str(line_dev))
```

```python
for i in range(0,300):
  for line_test in test_data_trial[i]:
    with open('BN-AuthProf_Dataset/Exp_Dataset/Age/Test/{0}.txt'.format(i+1), "w+") as f:
      for line_test in test_data_trial[i]:
        f.write(str(line_test))
    with open('BN-AuthProf_Dataset/Exp_Dataset/Gender/Test/{0}.txt'.format(i+1), "w+") as f:
      for line_test in test_data_trial[i]:
        f.write(str(line_test))
```
